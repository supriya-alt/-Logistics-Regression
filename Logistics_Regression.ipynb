#Question 6: What is Logistic Regression? How does it differ from Linear Regression?

->Logistic Regression is a statistical method used to predict categorical outcomes, especially binary outcomes (e.g., Yes/No, 0/1, Pass/Fail).
Instead of fitting a straight line, it models the probability that a given input belongs to a particular category using the logistic (sigmoid) function.

#Difference between Linear and Logistic Regression:

The main difference between Linear Regression and Logistic Regression lies in the type of problem they solve. Linear Regression is used for regression problems, where the target variable is continuous, such as predicting house prices, sales amounts, or student scores. In contrast, Logistic Regression is used for classification problems, where the target variable is categorical, such as predicting whether an email is spam or not, or whether a patient has a disease (Yes/No).

Another difference is in the equation used. Linear Regression follows the equation

Linear Regression models a straight-line relationship between the dependent and independent variables, assuming the relationship is linear. Logistic Regression does not assume a linear relationship between X and Y directly; instead, it models a non-linear S-shaped (sigmoid) curve through the log-odds of the dependent variable.

Finally, Linear Regression is mainly used when we are interested in predicting numeric outcomes, while Logistic Regression is used when we want to predict class membership or probability of a certain event occurring.

#Question 8: What is the purpose of the R-squared metric in regression analysis?

->R-squared (RÂ²), also known as the coefficient of determination, is a statistical metric used to measure how well a regression model explains the variation in the dependent variable based on the independent variables.

It indicates the goodness of fit of the model. The value of RÂ² ranges from 0 to 1:

RÂ² = 0 means the model does not explain any variation in the dependent variable.

RÂ² = 1 means the model perfectly explains all the variation in the dependent variable.

For example, an RÂ² value of 0.85 means that 85% of the variation in the target variable can be explained by the model, while the remaining 15% is due to other factors or random noise.

#Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.

->Below is the Python code to fit a Simple Linear Regression model using the scikit-learn library and display the slope and intercept of the regression line



```python
# Import required libraries
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)   # Independent variable
Y = np.array([2, 4, 5, 4, 5])                  # Dependent variable

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Print slope and intercept
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```
